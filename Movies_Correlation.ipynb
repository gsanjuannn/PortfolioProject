```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the config of the plots created


#read data
df = pd.read_csv(r'E:\Data\movies.csv')
```

```python
#show data
df.head()
```

```python
#check for missing data

for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pctMissing))
```

```python
#check data types for each column

df.dtypes
```

```python
#changing data tpyes

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#corrected year column

df['yearcorrect'] = df['released'].astype(str).str[:4]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#shows all the rows when displaying data
pd.set_option('display.max_rows', None)
```

```python
#check for duplicates
#df['company'].drop_duplicates().sort_values(ascending=False) #shows distinct values under 'company' column

df.drop_duplicates()
df.head()
```

```python
#Hypotheses
#Budget high correlation, more money they spend the more money  they'll make
#Company high correlation
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('budget vs gross')
plt.xlabel('budget (in hundred millions)')
plt.ylabel('gross (in hundred millinos)')
plt.show()
```

```python
df.head()
```

```python
#plot budget vs gross using seaborne
sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
#look at correlation
#types of correlation, methods
#pearson(default), kendall, spearman
df.corr(method='pearson')
```

```python
#visualize correaltions
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Move Features')
plt.ylabel('Move Features')


plt.show()
```

```python
#assign numerical values to 'company' str values

df_numerize = df
for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype == 'object'):
        df_numerize[col_name] = df_numerize[col_name].astype('category')
        df_numerize[col_name] = df_numerize[col_name].cat.codes

df_numerize.head()
```

```python
df.head()
```

```python
#visualize correaltions
correlation_matrix = df_numerize.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Move Features')
plt.ylabel('Move Features')


plt.show()
```

```python
df_numerize.corr()
```

```python
#unstacking

correlation_mat = df_numerize.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_correlations = sorted_pairs[(sorted_pairs) > 0.5]

high_correlations #votes and budget have the highest correlation to gross
```
